```python
sentences = [
    "I am studying Machine Learning this semester.",
    "I completed the Networking course last year.",

    "Tiger is the national animal of India.",
    "India has a rich cultural heritage.",
    "Rajasthan is a famous state in India.",

    "I enjoy playing football with my friends.",
    "I like watching basketball matches."
]
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer #TF-BoW
from sklearn.feature_extraction.text import TfidfVectorizer #TF-IDF
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#TF-BowEmbedding
tf_vectorizer=CountVectorizer()
tf_emb=tf_vectorizer.fit_transform(sentences)
```

```python
#similarity matrix using cosine_similarity
tf_sim=cosine_similarity(tf_emb)
```

```python
tf_sim
```

```python
#TF-BowEmbedding
tfidf_vectorizer=TfidfVectorizer()
tfidf_emb=tfidf_vectorizer.fit_transform(sentences)
```

```python
tfidf_sim=cosine_similarity(tfidf_emb)
```

```python
tfidf_sim
```

```python
#Recommend the most matching/similar sentence
def doc_sim(tfidf_sim,sentence):
  for i in range (len(sentences)):
    sim=tfidf_sim[i].copy()
    sim[i]=-1
    top_index=np.argmax(sim)
    print(sentences[i],"->",sentences[top_index])
```

```python
doc_sim(tfidf_sim,sentences)
```

```python
#SBERT for meaning level similarities

from sentence_transformers import SentenceTransformer
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')

sbert_embeddings = sbert_model.encode(sentences)
sbert_sim = cosine_similarity(sbert_embeddings)

print("SBERT Similarity Matrix:\n")
print(sbert_sim)

def sbert_doc_sim(sim_matrix, sentences):
    print("\nSBERT Semantic Recommendations:\n")
    for i in range(len(sentences)):
        sim = sim_matrix[i].copy()
        sim[i] = -1
        top_index = np.argmax(sim)
        print(sentences[i], "->", sentences[top_index])


sbert_doc_sim(sbert_sim, sentences)
```
