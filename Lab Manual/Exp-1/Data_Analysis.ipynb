```python
# Importing the Necessary Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1.Load a dataset in your IDE

dataset = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
# 2.Observe the statistics of all the features

dataset.describe()
```

```python
# 3.Obtain the shape of the dataset

dataset.shape
```

```python
# 4.Separate all the features

dataset.info()
```

```python
# Drop rows where target 'Survived' is missing (none in this dataset, but just in case)
dataset = dataset.dropna(subset=["Survived"])

# Fill missing values properly (no inplace warning)
dataset["Age"] = dataset["Age"].fillna(dataset["Age"].median())
dataset["Embarked"] = dataset["Embarked"].fillna(dataset["Embarked"].mode()[0])
dataset["Fare"] = dataset["Fare"].fillna(dataset["Fare"].median())

# Drop Cabin (too many missing values)
dataset = dataset.drop(columns=["Cabin"])

```

```python
# 6.Observe the Box-Plot of each feature

# For Rating

plt.boxplot(dataset['Age'])
```

```python
# For Size

plt.boxplot(dataset['Fare'])
```
