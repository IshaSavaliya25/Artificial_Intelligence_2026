```python
import nltk
from nltk.util import ngrams
from collections import defaultdict, Counter
import random
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Training corpus
corpus = """
I like machine learning
I like deep learning
I enjoy machine learning
I enjoy data science
Machine learning is interesting
Deep learning is powerful
"""
```

```python
# Tokenize sentences
tokens = nltk.word_tokenize(corpus.lower())
```

```python
# Build Bigram Model
n = 2
bigrams = list(ngrams(tokens, n))
```

```python
# Create probability table
model = defaultdict(Counter)
for w1, w2 in bigrams:
    model[w1][w2] += 1
```

```python
prob_model = {}
```

```python
for w1 in model:
    total_count = sum(model[w1].values())
    prob_model[w1] = {}
    for w2 in model[w1]:
        prob_model[w1][w2] = model[w1][w2] / total_count
```

```python
# Function to predict next word
def predict_next(word):
    if word in prob_model:
        next_words = prob_model[word]
        predicted_word = max(next_words, key=next_words.get)
        return predicted_word, next_words
    else:
        return "No prediction available", {}
```

```python
# Example prediction
input_word = "enjoy"
prediction, probabilities = predict_next(input_word)
print("Input:", input_word)
print("Predicted next word:", prediction)
print("Probability Distribution:", probabilities)
```
